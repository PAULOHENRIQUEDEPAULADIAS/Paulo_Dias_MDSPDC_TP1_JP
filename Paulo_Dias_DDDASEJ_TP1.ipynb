1. Explique de forma sucinta o que são microsserviços.

**R.1:** Microserviços é um tipo de arquitetura orientado a serviços, modelados com foco no domínio do negócio. Consiste em uma série de serviços com deploy independentes e possuem seus próprios recursos, também independentes (Como banco de dados e serviços de API). Se comunicam através de networks e possuem como característica principal sua independência, permitindo assim que cada time atue em um microserviço distinto sem afetar seus pares.

2. Cite uma vantagem da arquitetura de microsserviços.

**R.2:** Algumas vantagens da arquitetura de microserviços:

* Microserviços divide o negócio em serviços menores, possibilitando assim, diferentes times atuarem em seus respectivos serviços, sem afetar seus pares, com baixo/somente o necessário acoplamento entre serviços;
* Microserviços possui facilitade de escalabilidade onde, ao contrário de outros tipos de arquiteturas, este modelo escala somente o serviço necessário, garantindo economia de recursos em relação a serviços que não possuem necessidade de escalonamento.
* Como a arquitetura de microserviços é uma ferramenta que atua conforme a funcionalidade/serviço esperado, isso permite que cada time "tome propriedade" de seus próprios dados, garantindo maior coesão para o negócio global.
* Com isolamento de times por serviços, se faz possivel o uso de diferentes tecnologias, linguagens de programação, e locais de deploy, garantindo maior flexibilidade em termos de desenvolvimento (Em teoria).

3. Cite uma desvantagem da arquitetura de microsserviços.

**R.3:** Algumas desvantagens da arquitetura de microserviços:
* Comunicação via network, o que muitas vezes pode tornar a comicação entre os serviços mais latente do que desejamos;
* Caso um sistema escale em níveis inesperadamente grandes, isso pode resultar perca de qualidade e segurança, afetando validações que tornam o sistema mais confiável.
* A constante crescente em tecnologias que nos ajudam a implementar microserviços, também aumenta a proporção de possíveis falhas, o que tornam a necessidade de conhecimento e entendimento muito mais enfatizados.

4. Cite uma característica dos microsserviços que os diferenciam de outras arquiteturas de software.

**R.4:**
* Uma característica marcante dos microsserviços é o foco no domínio de negócio. Essa arquitetura organiza o sistema com base nas áreas funcionais da empresa, permitindo que cada serviço seja desenvolvido, implantado e mantido de forma independente. Diferente de outras arquiteturas que costumam se orientar mais pela tecnologia utilizada, os microsserviços priorizam o alinhamento com as necessidades e regras de negócio.

5. Explique de forma sucinta o que é um monólito.

**R.5:** Monolito trata-se de uma arquitetura onde o desenvolvimento é único e integrado, com um único deploy, muitas vezes com nível de acoplamento alto e baixa fleximilidade a alterações. Monolito é uma arquitetura ainda muito válida. Existem três tipos de monolito: O sistema de processo único; Um monolito distribuido e; O sistema black-box de terceiros.

6. Explique de forma sucinta o que significa Acoplamento do ponto de vista de Engenharia de Software.

**R.6:** Acoplamento em engenharia de software, indica a forte dependência e conexão entre funcionalidades do sistema. Exemplificando um acoplamento por compartilhamento de banco de dados é forte; comunicação via API bem definida é mais fraca. Quanto maior o nível de acoplamento, menor será a tolerância a mudanças, onde alterações mínimas requerirão ajustes em várias partes da aplicação.

7. Explique de forma sucinta o que significa Coesão do ponto de vista de Engenharia de Software.

**R.7:** Coesão em engenharia de software, indica o agrupamento de códigos similares, significa o agrupamento de funcionalidades que servem a um propósito em comum.

8. Modele um Value Object no padrão da UML contendo estereótipo, propriedades e métodos. Coloque um comentário no construtor informando que tipo de validações serão feitas.

**R.8:** Value Object se trata de um objeto imutável, ao qual não possui identificador e seu valor é definido por suas propriedades. Um exemplo de Value Object seria a classe endereo, ao qual é definida pelo endereço que contém.



9. Modele uma Entity no padrão da UML que use o Value Object da questão 5. **OBS: Creio que se trata do V.O. da questão 8**

**R.9:** Uma entidade se trata de uma classe, neste caso uma classe que use a classe Endereço (Value Object Adress) e a tenha como atributo complexo (Atribuindo seu tipo a classe desejada, neste caso, Endereço). Diferente de Value Object, Entity requer um identificador único e seu valor é identificado por este.



10. Porque o compartilhamento de banco de dados é uma estratégia de integração ruim sob o ponto de vista de microsserviços?

**R.10:** Em microserviços, não é o ideal mas pode acontecer da integração entre bancos. É considerado uma estratégia ruim pois acopla dois ou mais serviços, assim sendo, funcionalidades que precisam ser adaptadas para um serviço específico, muitas vezes também serão necessárias a adaptação para os outros. Uma estratégia para burlar essa necessidade, é a requisição dos dados para o serviço detentor do banco de dados, assim este pode filtrar informações, e aplicar suas regras específicas.

11. Explique de forma sucinta o que é um Agregado do DDD.

**R.11:** Um agregado representa um conjunto coerente de entidades e objetos de valor que formam uma unidade de consistência dentro do domínio. Ele possui um ponto de controle principal chamado agregado raiz. Por exemplo, em um domínio de vendas, o Pedido seria o agregado raiz, enquanto Item do Pedido, Endereço de Entrega e Forma de Pagamento seriam partes que compõem esse agregado.

12. Cite uma vantagem de construir microsserviços usando Agregados do DDD.

**R.12:** Uma vantagem de utilizar agregados na construção de microserviços é o isolamento de contexto e consistência interna. Cada agregado define limites claros de dados e regras de negócio, permitindo que o microserviço seja autônomo e coeso, possuindo todos os componentes necessários para o seu funcionamento sem depender diretamente de outros serviços.
