O Spring Boot e o Spring Cloud são ambos projetos do ecossistema Spring que se complementam. 

Spring Boot: possui a funcionalidade de facilitar a criação de aplicações Java, com uma série de recursos automáticos e atuam tanto em arquiteturas monolíticas ou baseadas em microserviços.

Spring Cloud: volta seu foco para comunicação entre serviços em arquiteturas baseadas em microserviços, como descoberta de serviços, balanceamento e distribuição de carga, configurações centralizadas, entre outras funcionalidades que facilitam a configuração de conexões para aplicações distribuídas.

Ambos possuem papéis complementares: o Spring Boot é utilizado em toda a aplicação desde a sua concepção, já que cada serviço é construído sobre ele. O Spring Cloud, por sua vez, entra em cena quando há necessidade de integração e coordenação entre esses serviços.

Arquiteturas monolíticas são criadas como uma implementação e deploy únicos, em que seus recursos são, geralmente, fortemente acoplados e dependentes entre si. Foram — e ainda são — amplamente utilizadas no desenvolvimento de aplicações web, especialmente aquelas que não possuem expectativa de grande crescimento ou cujo objetivo atual é atender a um público inicial reduzido e controlado. Em geral, esse modelo oferece limitações quanto à flexibilidade e ao escalonamento, o que pode afetar sua manutenção à medida que se torna mais complexo.

Arquiteturas baseadas em microserviços, por sua vez, solucionam alguns desses problemas. Intimamente relacionadas ao Domain-Driven Design (DDD), essas abordagens dividem o processo de desenvolvimento em serviços menores, possibilitando melhor controle e tornando cada serviço isolado — ou fracamente acoplado — em relação aos demais. Isso proporciona maior flexibilidade para atualizações e correções. Essa estratégia também permite que a equipe responsável por um serviço específico tenha total controle sobre seus próprios dados, tecnologias utilizadas, exposição de endpoints, entre outros benefícios. Atualmente, arquiteturas baseadas em microserviços são amplamente adotadas em plataformas de streaming, e-commerce e serviços financeiros, permitindo que as empresas mantenham controle isolado e flexível sobre cada serviço, aplicando alterações de forma robusta e ágil, com maior volume de entregas e melhor acompanhamento do mercado.

As arquiteturas baseadas em microserviços podem adotar diferentes padrões de implementação, como event-driven, API gateway, service mesh e CQRS, dependendo da necessidade de comunicação, escalabilidade e controle transacional.

Apesar de a arquitetura de microserviços atender a uma ampla gama de objetivos, ela nem sempre é a escolha ideal. Sua natureza distribuída pode representar um desafio em cenários que exigem o menor nível possível de latência — fator variável e imprevisível que oscila conforme o número de serviços na aplicação, já que a comunicação entre eles ocorre via rede (network). Além disso, sua adoção pode ser desnecessária em aplicações de baixa complexidade, onde a simplicidade de uma arquitetura monolítica é suficiente.

Atualmente, as aplicações estão cada vez mais sendo implementadas em serviços de nuvem, aproveitando vantagens como escalabilidade automatizada e ausência de necessidade de servidores próprios. Para alcançar esse objetivo de hosting, alguns princípios devem ser seguidos:

1. Design com foco em automação: projetar e estruturar componentes visando facilitar automações futuras — incluindo infraestrutura, atualizações, integração e entrega contínua (CI/CD), build, testing e deployment automatizados, além de escalonamento automático (up e down).
2. Seja inteligente com estados: definir cuidadosamente onde e quando armazenar estados da aplicação (ex.: dados de carrinho de compras, número de empregados, instâncias em execução, versões de código em produção). Sempre que possível, adote componentes stateless, pois são mais simples de escalar, reparar, fazer rollbacks e balancear.
3. Favoreça serviços gerenciados: priorize tecnologias que ofereçam suporte nativo e reduzam a necessidade de gerenciamento manual do backend.
4. Pratique defesa em profundidade: implemente múltiplas camadas de segurança entre os serviços, garantindo proteção adicional além das barreiras principais, prevenindo possíveis ataques.
5. Esteja sempre arquitetando: mantenha-se atualizado sobre tendências de mercado e implemente novos — ou complementares — modelos de arquitetura que se alinhem ao domínio do negócio e aos objetivos da organização.

Antes de iniciar o desenvolvimento de uma aplicação baseada em microserviços, é fundamental compreender os 12 princípios de desenvolvimento (inspirados nos Twelve-Factor Apps), que também se aplicam a outros tipos de arquitetura. Em ambientes distribuídos, contudo, a atenção a esses fatores deve ser redobrada:

* Base de código — Uma única base de código rastreada por controle de versão, com múltiplos deploys.
* Dependências — Declare e isole as dependências.
* Configurações — Armazene as configurações no ambiente.
*  Serviços de apoio — Trate serviços de apoio como recursos anexados.
* Build, release, run — Separe estritamente os estágios de build, release e execução.
* Processos — Execute a aplicação como um ou mais processos stateless.
* Vínculo de porta — Exporte serviços via ligação de porta.
* Concorrência — Escalone por meio de processos.
* Descartabilidade — Maximize a robustez com inicialização e desligamento rápidos.
* Ambientes dev/prod semelhantes — Mantenha desenvolvimento, teste e produção o mais semelhantes possível.
* Logs — Trate logs como fluxos de eventos.
* Processos administrativos — Execute tarefas administrativas ou de manutenção como processos pontuais.

Em resumo, tanto aplicações com arquitetura monolítica quanto baseadas em microserviços possuem suas vantagens e são indicadas para cenários específicos. Arquiteturas monolíticas são preferíveis em aplicações simples, de baixa complexidade e com projeção de acesso controlado. Já os microserviços são recomendados para aplicações complexas, com grande volume de acessos, múltiplos domínios funcionais (como autenticação, segurança e escalabilidade horizontal) e que se beneficiam de um processo de manutenção e desenvolvimento mais modular.

### Resolução

Para esta aplicação, desenvolvida com base na arquitetura de microsserviços do tipo REST, foram utilizados os seguintes serviços:

* service_discovery – servidor → atua com a abertura da porta 8761, configurada com a biblioteca Eureka;

* hello_api – método GET /hello → retorna a mensagem "Hello ";

* world_api – método GET /world → retorna a mensagem "World!";

* hello_world_api – método GET /world → integra os endpoints hello_api e world_api, retornando a mensagem "Hello World!";

* gateway_api – portão de gerenciamento de rotas → atua no gerenciamento das rotas registradas no servidor Eureka, funcionando como intermediário e facilitando as requisições.

Para o serviço service_discovery, o arquivo application.properties foi configurado da seguinte forma:

* Porta do servidor: 8761;

* O service_discovery não se autoregistra no servidor como uma API, nem tenta localizar a si mesmo dentro dele, evitando assim recursão e possíveis falhas.

Para que ele possa atuar como servidor, foi utilizada a anotação @EnableEurekaServer dentro da classe ServiceDiscoveryApplication.

Para o serviço gateway_api, a configuração foi realizada da seguinte forma para que ele possa efetuar a filtragem necessária:

* Porta de acesso: 8080;

* Definição do endereço onde o servidor está localizado: porta 8761/eureka/;

* O gateway está habilitado para realizar buscas e se registrar no servidor;

* Por fim, a propriedade discovery.locator permite que ele realize buscas dentro do servidor e que essas buscas sejam registradas em letras minúsculas (lowercase).

Tanto a hello_api quanto a world_api seguem o mesmo formato. Para exemplificação, será utilizada a hello_api.
No caso da hello_api, o arquivo application.properties está configurado da seguinte forma:

* Define a porta de acesso da API: 8081;

* Especifica a localização do servidor: 8761/eureka/;

* Permite que a aplicação se registre no servidor e realize requisições.

Por se tratar de uma aplicação simples, dentro do arquivo HelloApiApplication foi definido um controlador com as anotações @RestController e @RequestMapping("hello").
Esse controlador possui o método hello, anotado com @GetMapping, que retorna a string "Hello ".

Por fim, para a hello_world_api, o arquivo application.properties foi configurado exatamente da mesma forma que o correspondente da hello_api, e o arquivo HelloWorldApiApplication foi estruturado da seguinte maneira:

* Possui uma classe HelloWorld, anotada com @RestController e @RequestMapping("hello_world");

* Para facilitar as requisições aos dois serviços anteriores (hello_api e world_api), é declarado o DiscoveryClient com a anotação @Autowired, e, para a harmonização do resultado, é criada uma instância de RestTemplate;

* Contém um método hello_world, anotado com @GetMapping, que retorna a combinação das respostas de hello_api e world_api.

Para o funcionamento adequado, execute os serviços na seguinte ordem:

1. service_discovery;

2. hello_api, world_api e hello_world_api;

3. gateway_api.
